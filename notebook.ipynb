```python
import matplotlib.pyplot as plt
from orbit_generator import generate_orbits_dataset
```

```python
df = generate_orbits_dataset(n_orbits=10, two_d=True, num_steps=100)
```

```python
# Simple plot of first few orbits in ECI
fig, ax = plt.subplots(figsize=(10, 10))
unique_ids = df["orbit_id"].unique()
for oid in unique_ids:
    sub_df = df[df["orbit_id"] == oid]
    ax.plot(sub_df["x_eci_km"], sub_df["y_eci_km"], label=oid)
earth = plt.Circle((0, 0), 6371, color='blue', fill=True)
ax.add_artist(earth)
ax.set_aspect('equal')
ax.set_xlim(-50000, 50000)
ax.set_ylim(-50000, 50000)
ax.set_xlabel('X ECI (km)')
ax.set_ylabel('Y ECI (km)')
plt.title('Sample 2D Planar Orbits')
plt.legend()
plt.grid(True)
plt.show()
```
